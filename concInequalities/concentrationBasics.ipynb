```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Parameters
true_mean = 50
true_std = 10
sample_size = 30
num_samples = 1_000_000

# Generate data: take 1000 sample means, each of size 30
sample_means = [np.mean(np.random.normal(true_mean, true_std, sample_size)) for _ in range(num_samples)]

# Convert to numpy array
sample_means = np.array(sample_means)

# Plotting
plt.figure(figsize=(16, 12))

# Histogram of sample means
plt.subplot(2, 2, 1)
plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='blue', edgecolor='black')
plt.axvline(true_mean, color='red', linestyle='dashed', linewidth=2, label=f"True Mean = {true_mean}")
plt.title("Histogram of Sample Means")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.legend()

# Boxplot of sample means
plt.subplot(2, 2, 2)
plt.boxplot(sample_means, vert=False)
plt.axvline(true_mean, color='red', linestyle='dashed', linewidth=2, label="True Mean")
plt.title("Boxplot of Sample Means")
plt.xlabel("Sample Mean")
plt.legend()

# Convergence of sample means over iterations
plt.subplot(2, 2, 3)
plt.plot(sample_means, marker='o', markersize=3, alpha=0.5, linestyle='None')
plt.axhline(true_mean, color='red', linestyle='dashed', linewidth=2, label="True Mean")
plt.title("Sample Means across 1000 Samples")
plt.xlabel("Sample Index")
plt.ylabel("Sample Mean")
plt.legend()

# Distribution of original population vs sample means
x = np.linspace(true_mean - 4*true_std, true_mean + 4*true_std, 200)
pop_dist = 1/(true_std*np.sqrt(2*np.pi)) * np.exp(-(x-true_mean)**2/(2*true_std**2))
sm_std = true_std / np.sqrt(sample_size)
sm_dist = 1/(sm_std*np.sqrt(2*np.pi)) * np.exp(-(x-true_mean)**2/(2*sm_std**2))

plt.subplot(2, 2, 4)
plt.plot(x, pop_dist, label="Population Distribution", color="green")
plt.plot(x, sm_dist, label="Sampling Distribution", color="blue")
plt.axvline(true_mean, color='red', linestyle='dashed', linewidth=2)
plt.title("Population vs Sampling Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()

plt.tight_layout()
plt.show()
```
